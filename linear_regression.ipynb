## **Importing the libraries**

```
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## **Allocating the Data**

180 , 10

154, 6



```
X = [[180],[154],[155],[178],[186],[168],[178],[170],[172],[173],[174],[188]]
Y = [9,6,7,10,10,8,9,9,8,9,9,13]
print(X)
print(Y)
```

## **Plotting our data**

```
plt.scatter(X,Y)
plt.show()
```

## **Fitting the data to the model**

```
clf = LinearRegression()
clf.fit(X,Y)
print(clf.score(X,Y))
```

## **Predicting**

```
print(clf.predict([[168]]))
```

## **the fit line**

```
plt.scatter(X,Y)
plt.plot(X,clf.predict(X),color='red')
plt.show()
```
